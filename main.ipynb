Так как нужна визуализация между нашим таргетом(адресатом) и атрибутов то обычные графики boxplot и так далее тут просто так не построишь так как текст лучше будет наши текста преобразовать в числа и построить график на основе них

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
count_vectorizer = CountVectorizer(max_features=500)
tfidf_vectorizer = TfidfVectorizer(max_features=500)
X_tfidf = tfidf_vectorizer.fit_transform(df_drop_new['Краткое\nсодержание'])
X_count = count_vectorizer.fit_transform(df_drop_new['Краткое\nсодержание'])
tsne_count = TSNE(n_components=2, random_state=42)
X_count_tsne = tsne_count.fit_transform(X_count.toarray())
X_tfidf_tsne = tsne_count.fit_transform(X_tfidf.toarray())

df_visualization_count = pd.DataFrame({
    'X': X_count_tsne[:, 0],
    'Y': X_count_tsne[:, 1],
    'Адресат': df_drop_new['Адресат']
})

df_visualization_tfidf = pd.DataFrame({
    'X': X_tfidf_tsne[:, 0],
    'Y': X_tfidf_tsne[:, 1],
    'Адресат': df_drop_new['Адресат']
})

plt.figure(figsize=(12, 8))
sns.scatterplot(x='X', y='Y', hue='Адресат', data=df_visualization_count, palette='viridis', markers=True)
plt.title('Визуализация с использованием CountVectorizer')
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='X', y='Y', hue='Адресат', data=df_visualization_tfidf, palette='viridis', markers=True)
plt.title('Визуализация с использованием TfidfVectorizer')
plt.legend(loc='upper right')
plt.show()
```

```python
import pymorphy3
from nltk.tokenize import word_tokenize
import re
def preproc_text(df):
    text=df['Краткое\nсодержание']
    norm_text=[]
    morph=pymorphy3.MorphAnalyzer()
    for s in text:
        s1 = re.sub(r'[^\w\s]+|[\d]+', r'',s).strip()
        s1 = s1.lower()
        s1 = word_tokenize(s1)
        words=[]
        for i in s1:
            pv = morph.parse(i)
            words.append(pv[0].normal_form)
        sentence=' '.join(words)
        norm_text.append(sentence)
    return norm_text
```
